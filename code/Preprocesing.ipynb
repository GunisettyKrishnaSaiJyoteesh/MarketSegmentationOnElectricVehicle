```python
  # Age and Salary are exact values

# Step 1: Extract numeric ranges from df
df[['Age_min', 'Age_max']] = df['Age'].str.split('-', expand=True).astype(int)
df[['Income_min', 'Income_max']] = df['Income'].str.replace(' ', '').str.split('-', expand=True).astype(int)

# Step 2: Add temp keys for cross join
df['_key'] = 1
df2['_key'] = 1

# Step 3: Cross join
df_cross = pd.merge(df, df2, on='_key').drop(columns=['_key'])

# Step 4: Filter rows that match both Age and Salary within ranges
matched = df_cross[
    (df_cross['Age_y'] >= df_cross['Age_min']) & 
    (df_cross['Age_y'] <= df_cross['Age_max']) &
    (df_cross['Salary'] >= df_cross['Income_min']) & 
    (df_cross['Salary'] <= df_cross['Income_max'])
].copy()

# Step 5: Add unique identifiers to original dataframes
df['_id1'] = df.index
df2['_id2'] = df2.index

# Step 6: Track matched IDs
matched_ids_df1 = matched['Age_min'].index.unique()
matched_ids_df2 = matched['Age_y'].index.unique()

# Step 7: Unmatched entries
used_df1_rows = matched[['Age_min', 'Income_min', 'Age_max', 'Income_max']].drop_duplicates()
unmatched_df1 = df.merge(used_df1_rows, on=['Age_min', 'Income_min', 'Age_max', 'Income_max'], how='outer', indicator=True)
unmatched_df1 = unmatched_df1[unmatched_df1['_merge'] == 'left_only'].drop(columns=['_merge'])

used_df2_rows = matched[['Age_y', 'Salary']].drop_duplicates()
unmatched_df2 = df2.merge(used_df2_rows, left_on=['Age', 'Salary'], right_on=['Age_y', 'Salary'], how='outer', indicator=True)
unmatched_df2 = unmatched_df2[unmatched_df2['_merge'] == 'left_only'].drop(columns=['_merge', 'Age_y'])

# Step 8: Concatenate all
matched['source'] = 'matched'
unmatched_df1['source'] = 'df1_only'
unmatched_df2['source'] = 'df2_only'

# Optional: align column names to be consistent
final_cols = matched.columns.union(unmatched_df1.columns).union(unmatched_df2.columns)
matched = matched.reindex(columns=final_cols)
unmatched_df1 = unmatched_df1.reindex(columns=final_cols)
unmatched_df2 = unmatched_df2.reindex(columns=final_cols)

# Final output
final_df = pd.concat([matched, unmatched_df1, unmatched_df2], ignore_index=True)

# Save or view

```

```python
df = pd.read_csv("final_output.csv")

# -------------------------------
# 2) Identify numeric vs. categorical columns
# -------------------------------
numeric_cols = df.select_dtypes(include=[np.number]).columns.tolist()
categorical_cols = df.select_dtypes(
    include=["object", "category"]
).columns.tolist()
```

```python
from sklearn.impute import KNNImputer, SimpleImputer
from sklearn.preprocessing import LabelEncoder
from sklearn.cluster import KMeans
```

```python
if len(categorical_cols) > 0:
    cat_imputer = SimpleImputer(strategy="most_frequent")
    df[categorical_cols] = cat_imputer.fit_transform(df[categorical_cols])


if len(numeric_cols) > 0:
    # You can tune n_neighbors as you see fit.
    knn_imp = KNNImputer(n_neighbors=5, weights="distance")
    df[numeric_cols] = knn_imp.fit_transform(df[numeric_cols])

df.to_csv("final_output_knnimputed.csv", index=False)



def kmeans_impute(df_original, numeric_cols, n_clusters=5, random_state=0):
    """
    Returns a copy of df_original where numeric columns are imputed via K-Means cluster centroids.
    Categorical columns are assumed to have already been imputed or encoded.
    """
    df = df_original.copy()
    num_df = df[numeric_cols]

    # 4a) Temporary fill: use column means so that KMeans can run
    temp_fill = num_df.fillna(num_df.mean())

    # 4b) Fit KMeans on the fully‐filled data
    kmeans = KMeans(n_clusters=n_clusters, random_state=random_state)
    kmeans.fit(temp_fill)

    # 4c) For each row with any NaNs in numeric_cols, replace missing by cluster center
    centers = kmeans.cluster_centers_  # shape (n_clusters, n_numeric_features)

    # We will assign each row to its closest cluster (using the temporarily filled values)
    labels = kmeans.labels_  # each row’s cluster index

    # Now, for rows that originally had NaN in numeric_cols, overwrite:
    for idx, row in num_df.iterrows():
        if row.isna().any():
            cluster_id = labels[idx]
            # For each column that was NaN, set to the cluster’s centroid value
            for j, col in enumerate(numeric_cols):
                if pd.isna(row[col]):
                    df.at[idx, col] = centers[cluster_id, j]

    return df

df_kmeans_imputed = kmeans_impute(df, numeric_cols, n_clusters=5, random_state=42)
df_kmeans_imputed.to_csv("final_output_kmeans_imputed.csv", index=False)
```
